# 追踪手部

该部分先**导入了检测手部的训练好的模型**

如果在视频中检测出了手部则直接返回**对应的坐标** or 用**点**来表示已侦测到了

````python
# 导入库
import cv2
import mediapipe as mp

#读取视频信息
cap = cv2.VideoCapture(0)

#导入检测手部的库
mpHands = mp.solutions.hands

hands = mpHands.Hands()


#如果读取成功则输出视频
while True:
    ret, imgBGR = cap.read()
    # 读不到就放弃
    if not ret:
        break;
    
    # 将bgr图片转换为rgb图片（从cap得到的默认为bgr图片）
    imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
    # 找到手部
    result = hands.process(imgRGB)
    
    # 侦测到手部的时候给个信息
    if result.multi_hand_landmarks:
        #回传所有手的坐标
        print(result.multi_hand_landmarks)
        print(".",end=" ")
    
    #给人看还是得BGR
    cv2.imshow('img',imgBGR)
    
    
    
    # 退出按'q'
    if cv2.waitKey(1)== ord('q'):
        print("Quit successfully")
        break

#好习惯


# 释放内存
cv2.destroyAllWindows()
# 释放摄像头
cap.release()
````

# 根据坐标画点上去

````python
# 导入库
import cv2
import mediapipe as mp

#读取视频信息
cap = cv2.VideoCapture(0)

#导入检测手部的库
mpHands = mp.solutions.hands

hands = mpHands.Hands()

# 导入由于根据坐标画点的函数
###############################
mpDraw = mp.solutions.drawing_utils



#如果读取成功则输出视频
while True:
    ret, imgBGR = cap.read()
    # 读不到就放弃
    if not ret:
        break;
    
    # 将bgr图片转换为rgb图片（从cap得到的默认为bgr图片）
    imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
    # 找到手部
    result = hands.process(imgRGB)
    
    # 侦测到手部的时候给个信息
    if result.multi_hand_landmarks:
        #回传所有手的坐标
        #print(result.multi_hand_landmarks)
        #print(".",end=" ")
        ##########################################
        #开始画点
        for handLms in result.multi_hand_landmarks:
            # 告诉系统话画哪张图上面，并且传入坐标
            # 遍历画上去所有点
            mpDraw.draw_landmarks(imgBGR,handLms)
        
    
    #给人看还是得BGR
    cv2.imshow('img',imgBGR)
    
    
    
    # 退出按'q'
    if cv2.waitKey(1)== ord('q'):
        print("Quit successfully")
        break

#好习惯
# 释放内存
cv2.destroyAllWindows()
# 释放摄像头
cap.release()
````

# 链接点

````python
# 导入库
import cv2
import mediapipe as mp

#读取视频信息
cap = cv2.VideoCapture(0)

#导入检测手部的库
mpHands = mp.solutions.hands

hands = mpHands.Hands()

# 导入由于根据坐标画点的函数
mpDraw = mp.solutions.drawing_utils

#####################################
# 更改点和线的设定
handLmsStyle = mpDraw.DrawingSpec(color=(0,0,255),thickness=5)
handConStyle = mpDraw.DrawingSpec(color=(0,255,0),thickness=10)


#如果读取成功则输出视频
while True:
    ret, imgBGR = cap.read()
    # 读不到就放弃
    if not ret:
        break;
    
    # 将bgr图片转换为rgb图片（从cap得到的默认为bgr图片）
    imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
    # 找到手部
    result = hands.process(imgRGB)
    
    # 侦测到手部的时候给个信息
    if result.multi_hand_landmarks:
        #回传所有手的坐标
        #print(result.multi_hand_landmarks)
        #对于找到的每一个手
        for handLms in result.multi_hand_landmarks:
            # 告诉系统话画哪张图上面，并且传入坐标
            # 第一个是画上去的图片，第二个是传入点位坐标，第三个是链接关节点
            #第四个是设定点的样式，第五个是设定线的样式（要在上面添加mpDrwa）
            #######################################
            mpDraw.draw_landmarks(imgBGR,handLms,mpHands.HAND_CONNECTIONS,handLmsStyle,handConStyle)
            #遍历的时候显示是第几个点,找到点的位置
            for i, lm in enumerate(handLms.landmark):
                print(i,lm.x,lm.y)
    
    #给人看还是得BGR
    cv2.imshow('img',imgBGR)
    
    
    
    # 退出按'q'
    if cv2.waitKey(1)== ord('q'):
        print("Quit successfully")
        break

#好习惯
# 释放内存
cv2.destroyAllWindows()
# 释放摄像头
cap.release()
````

# 获取每个点的坐标并添加（由比例乘以长宽）

````python
# 导入库
import cv2
import mediapipe as mp

#读取视频信息
cap = cv2.VideoCapture(0)

#导入检测手部的库
mpHands = mp.solutions.hands

hands = mpHands.Hands()

# 导入由于根据坐标画点的函数
mpDraw = mp.solutions.drawing_utils

# 更改点和线的设定
handLmsStyle = mpDraw.DrawingSpec(color=(0,0,255),thickness=5)
handConStyle = mpDraw.DrawingSpec(color=(0,255,0),thickness=10)


#如果读取成功则输出视频
while True:
    ret, imgBGR = cap.read()
    # 读不到就放弃
    if not ret:
        break;
    
    # 将bgr图片转换为rgb图片（从cap得到的默认为bgr图片）
    imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
    # 找到手部
    result = hands.process(imgRGB)
    
    # 侦测到手部的时候给个信息
    if result.multi_hand_landmarks:
        #回传所有手的坐标
        #print(result.multi_hand_landmarks)
        ##############################################
        # 找到改框框的屏幕长宽
        imgHeight = imgBGR.shape[0]
        imgWidth = imgBGR.shape[1]
        
        
        
        #对于找到的每一个手
        for handLms in result.multi_hand_landmarks:
            # 告诉系统话画哪张图上面，并且传入坐标
            # 第一个是画上去的图片，第二个是传入点位坐标，第三个是链接关节点
            #第四个是设定点的样式，第五个是设定线的样式（要在上面添加mpDrwa）
            mpDraw.draw_landmarks(imgBGR,handLms,mpHands.HAND_CONNECTIONS,handLmsStyle,handConStyle)
            #遍历的时候显示是第几个点,找到点的位置
            for i, lm in enumerate(handLms.landmark):
                ###################################
                #获得坐标后转成int好看些
                xPos = int(lm.x * imgWidth)
                yPos = int(lm.y * imgHeight)
                #将数字插入到图片里
                # (图片)，字符串，位置，字体，大小，颜色,粗度
                cv2.putText(imgBGR,str(i),(xPos-25,yPos+5),cv2.FONT_HERSHEY_SIMPLEX,0.4,(0,0,255),2)
                #print(i,xPos,yPos)
    
    #给人看还是得BGR
    cv2.imshow('img',imgBGR)
    
    
    
    # 退出按'q'
    if cv2.waitKey(1)== ord('q'):
        print("Quit successfully")
        break

#好习惯
# 释放内存
cv2.destroyAllWindows()
# 释放摄像头
cap.release()
````

# 加入帧数

需要多引入一个模组

```import time```

````python
# 导入库
import cv2
import mediapipe as mp
#########################
# 加入时间
import time


#读取视频信息
cap = cv2.VideoCapture(0)

#导入检测手部的库
mpHands = mp.solutions.hands

hands = mpHands.Hands()

# 导入由于根据坐标画点的函数
mpDraw = mp.solutions.drawing_utils

# 更改点和线的设定
handLmsStyle = mpDraw.DrawingSpec(color=(0,0,255),thickness=5)
handConStyle = mpDraw.DrawingSpec(color=(0,255,0),thickness=10)

####
#初始化pTime和cTime
pTime = 0
cTime = 0

#如果读取成功则输出视频
while True:
    ret, imgBGR = cap.read()
    # 读不到就放弃
    if not ret:
        break;
    
    # 将bgr图片转换为rgb图片（从cap得到的默认为bgr图片）
    imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
    # 找到手部
    result = hands.process(imgRGB)
    
    # 侦测到手部的时候给个信息
    if result.multi_hand_landmarks:
        #回传所有手的坐标
        #print(result.multi_hand_landmarks)
        # 找到改框框的屏幕长宽
        imgHeight = imgBGR.shape[0]
        imgWidth = imgBGR.shape[1]
        
        
        
        #对于找到的每一个手
        for handLms in result.multi_hand_landmarks:
            # 告诉系统话画哪张图上面，并且传入坐标
            # 第一个是画上去的图片，第二个是传入点位坐标，第三个是链接关节点
            #第四个是设定点的样式，第五个是设定线的样式（要在上面添加mpDrwa）
            mpDraw.draw_landmarks(imgBGR,handLms,mpHands.HAND_CONNECTIONS,handLmsStyle,handConStyle)
            #遍历的时候显示是第几个点,找到点的位置
            for i, lm in enumerate(handLms.landmark):
                #获得坐标后转成int好看些
                xPos = int(lm.x * imgWidth)
                yPos = int(lm.y * imgHeight)
                # 找大拇指的位置
                if i == 4: 
                # (图片)，字符串，位置，字体，大小，颜色,粗度
                    cv2.circle(imgBGR,(xPos,yPos),20,(166,56,56),cv2.FILLED)
                
    
    
    #############################
    # 加入帧数
    cTime = time.time()
    fps = 1/(cTime-pTime)
    pTime = cTime
    cv2.putText(imgBGR,f"FPS : {int(fps)}",(30,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),3)
    
    
    #给人看还是得BGR
    cv2.imshow('img',imgBGR)
    
    # 退出按'q'
    if cv2.waitKey(1)== ord('q'):
        print("Quit successfully")
        break

#好习惯
# 释放内存
cv2.destroyAllWindows()
# 释放摄像头
cap.release()
````

# 调参

对于 **mpHands.Hands()** 内参数的解释:



- 第一个是要你侦测静态or动态图片
    1. 动态为False
    2. 静态为True


- 第二个则是你最多会出现的手的数量


- 模型的复杂度（0是不精准，1是精准）
    1. 只能设定 0 or 1
    
- 侦测的置信度（超过多少概率时将其显示）

- 追踪的置信度（超过多少概率需要重新侦测手的位置）

````python
# 导入库
import cv2
import mediapipe as mp
# 加入时间
import time


#读取视频信息
cap = cv2.VideoCapture(0)

#导入检测手部的库
mpHands = mp.solutions.hands
##################################################
hands = mpHands.Hands(model_complexity=0,min_tracking_confidence=0.5)

# 导入由于根据坐标画点的函数
mpDraw = mp.solutions.drawing_utils

# 更改点和线的设定
handLmsStyle = mpDraw.DrawingSpec(color=(0,0,255),thickness=5)
handConStyle = mpDraw.DrawingSpec(color=(0,255,0),thickness=10)

####
#初始化pTime和cTime
pTime = 0
cTime = 0

#如果读取成功则输出视频
while True:
    ret, imgBGR = cap.read()
    imgBGR=cv2.flip(imgBGR, 1)
    # 读不到就放弃
    if not ret:
        break;
    
    # 将bgr图片转换为rgb图片（从cap得到的默认为bgr图片）
    imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
    # 找到手部
    result = hands.process(imgRGB)
    
    # 侦测到手部的时候给个信息
    if result.multi_hand_landmarks:
        #回传所有手的坐标
        #print(result.multi_hand_landmarks)
        # 找到改框框的屏幕长宽
        imgHeight = imgBGR.shape[0]
        imgWidth = imgBGR.shape[1]
        
        
        
        #对于找到的每一个手
        for handLms in result.multi_hand_landmarks:
            # 告诉系统话画哪张图上面，并且传入坐标
            # 第一个是画上去的图片，第二个是传入点位坐标，第三个是链接关节点
            #第四个是设定点的样式，第五个是设定线的样式（要在上面添加mpDrwa）
            mpDraw.draw_landmarks(imgBGR,handLms,mpHands.HAND_CONNECTIONS,handLmsStyle,handConStyle)
            #遍历的时候显示是第几个点,找到点的位置
            for i, lm in enumerate(handLms.landmark):
                #获得坐标后转成int好看些
                xPos = int(lm.x * imgWidth)
                yPos = int(lm.y * imgHeight)
                # 找大拇指的位置
                if i == 4: 
                # (图片)，字符串，位置，字体，大小，颜色,粗度
                    cv2.circle(imgBGR,(xPos,yPos),20,(166,56,56),cv2.FILLED)
                
    # 加入帧数
    cTime = time.time()
    fps = 1/(cTime-pTime)
    pTime = cTime
    cv2.putText(imgBGR,f"FPS : {int(fps)}",(30,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),3)
    
    
    #给人看还是得BGR
    cv2.imshow('img',imgBGR)
    
    # 退出按'q'
    if cv2.waitKey(1)== ord('q'):
        print("Quit successfully")
        break

#好习惯
# 释放内存
cv2.destroyAllWindows()
# 释放摄像头
cap.release()
````

# 结合人脸识别

````python
import cv2
import time
import mediapipe as mp

# 导入由于根据坐标画点的函数
mpDraw = mp.solutions.drawing_utils

#导入模型
mp_holistic = mp.solutions.holistic
holistic = mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5)

#导入视频
cap = cv2.VideoCapture(0)
# time.sleep(2)

# 更改点和线的设定
handLmsStyle = mpDraw.DrawingSpec(color=(0, 0, 255), thickness=5)
handConStyle = mpDraw.DrawingSpec(color=(0, 255, 0), thickness=10)


# 如果有信号则传入
while cap.isOpened():
    #读取数据
    success, imageBGR = cap.read()
    if not success:
        print("The vedio is not read successfully")
        continue
    
    # 将其由BGR转为RGB
    image = cv2.cvtColor(cv2.flip(imageBGR, 1), cv2.COLOR_BGR2RGB)
    
    results = holistic.process(image)
        
    mpDraw.draw_landmarks(imageBGR, results.face_landmarks, mp_holistic.FACEMESH_CONTOURS)
    mpDraw.draw_landmarks(imageBGR, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
    mpDraw.draw_landmarks(imageBGR, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
    mpDraw.draw_landmarks(imageBGR, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS)
        
    cv2.imshow('MediaPipe Holistic', imageBGR)
        
    # 退出按'q'
    if cv2.waitKey(1) == ord('q'):
        print("Quit successfully")
        break

# 释放内存
cv2.destroyAllWindows()
# 释放摄像头
cap.release()
````
