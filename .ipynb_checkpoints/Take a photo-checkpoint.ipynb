```python
# A Naive recursive Python implementation of LCS problem
import numpy as np 
def lcs(X, Y, m, n):
    comfirm=[[0 for k in range(n+1)] for l in range(m+1)]
    # value initially in each cell
    Comp = [[0 for k in range(n+1)] for l in range(m+1)]
    for i in range(m + 1):
        for j in range(n + 1):
            if j == 0 or i == 0:
                Comp[i][j] = 0
            elif X[i-1] == Y[j-1]:
                Comp[i][j] = 1 + Comp[i-1][j-1];
                comfirm[i][j]=1
            else:
                if Comp[i][j-1]>Comp[i-1][j]:
                    Comp[i][j]=Comp[i][j-1]
                else:
                    Comp[i][j]=Comp[i-1][j]
    
    display(Comp)
    display(comfirm)
    i=m
    j=n
    str_list=[]
    while(i>=0 and j>=0):
        if comfirm[i][j]==1:
            str_list.append(X[i-1])
            j=j-1
            i=i-1
            continue
        if Comp[i][j-1]>Comp[i-1][j]:
            j=j-1
        else:
            i=i-1
    
    #display(str_list)
    str_list.reverse()
    same_l = "".join(str_list)
    return Comp[m][n],same_l
    
# Driver program to test the above function
X = "ABCDGH"
Y = "AEDFHR"

re=lcs(X , Y, len(X), len(Y))
print("Length of LCS is ",re[0])
print(re[1])
```

```python
def Comp(A, B, m, n):
    # value initially in each cell
    Comp = [[0 for k in range(n+1)] for l in range(m+1)]
    pre=[["0" for k in range(n+1)] for l in range(m+1)]
    # To store the length of
    # longest common substring
    result = [0,0,0]
 
    # Following steps to build
    # Comp[m+1][n+1] in bottom up fashion
    for i in range(m + 1):
        for j in range(n + 1):
            if (i == 0 or j == 0):
                Comp[i][j] = 0
            elif (A[i-1] == B[j-1]):
                Comp[i][j] = Comp[i-1][j-1] + 1
                if Comp[i][j]>0:
                    pre[i][j] = A[i-1]
                if Comp[i][j]>result[0]:
                    result = [Comp[i][j],i,j]
            else:
                Comp[i][j] = 0
    a=result[1]
    b=result[2]
    same_l=[]
    count=0
    while(count<result[0]):
        count+=1
        same_l.append(pre[a][b])
        a-=1
        b-=1
    same_l.reverse()
    same_l = "".join(same_l)
    
    display(Comp)
    display(pre)
    return result[0],same_l
# Driver Code
X = 'abdesssssss'
Y = 'sasadabde'
m = len(X)
n = len(Y)
re=Comp(X, Y, m, n)
print('Length of Longest Common Substring is',re[0])
print(re[1])
```
